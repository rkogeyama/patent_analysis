## Introduction

Two main broad classes in ML algorithms
*  Penalized Linear Regression (eg. Ridge and Lasso)
    *  Improves OLS
    *  Top choice for predictive analytic
    *  Allows for balancing between overfitting and underfitting
    *  Offers information about the relative relevance of inputs
    *  Best prediction performance for underdetermined problems and problems with many inputs such as genetics and text mining.
    *  Recent developments of coordinate descent methods makes training penalized regression models extremely fast.
    
*  Ensemble Methods (eg. Random Forests and Gradient Boosting)
    *  Overdetermined models (eg. search results)
    *  Performance and ease of use
    *  Evolved from binary decisions trees
    

## Chapter 1
One class of machine learning problems, known as *function approximation*. 
Function approximation is a subset of *supervised learning* problems. OLS and logistic regression are examples of function approximation problems. 
Can be used for text classification, search responses, ad placements, spam filtering, predicting customer behavior, diagnostics, etc.

Two classes of algorithms for solving function approximation problems: Penalized linear regression methods and Ensemble methods. 

These classes provide optimum or near-optimum performance on the vast majority of predictive analytics (function approximation) problems

(evidence can be found in Rich Caruana's papers)

### Penalized regression models
Penalized regression models are easy to use, since they do not have many parameters, and they have well defined input types. Moreover, they indicate which of their input is most important for producing predictions. 

Penalized linear regression models were designed to overcome some limitations of OLS regression. The basic problem is overfitting. Problem: there are more variables than the number of data points (for example, in the case of genetic studies, or text mining, where there are thousands of word pairs). Penalized linear regression offer ways to systematically reduce the number of degrees of freedom.

### Ensemble models
The basic idea of ensemble models is to generate many predictive models and combine their output. One problem  of ensemble methods is their instability, the addition of data may cause radical changes in the output or performance. 

### How to choose the algorithm
_Training Time_ - Training can take days or weeks. Penalized methods are faster, and they might serve as a first step in a development process.

_Choosing inputs_ - Throwing in everything you have is not usually a good solution. It takes some iteration to select the relevant input.

_Feature extraction_ - Collect data in free form and transform to a database-like organization
_Feature engineering_ - Manipulating features (data) to generate more informative ones.
