```python
src='data/sample.csv'
# src='data/sample.csv'
```

```python
g=Graph()
```

```python
import csv
with open(src, 'r') as csvfile:
    
    list_of_edges = csv.reader(csvfile, delimiter=',')

    vertices = {}

    for e in list_of_edges:
        if e[0] not in vertices:
            vertices[e[0]] = True
        if e[1] not in vertices:
            vertices[e[1]] = True

    for d in vertices:
        vertices[d] = g.add_vertex()

    for edge in list_of_edges:
        g.add_edge(vertices[edge[0]], vertices[edge[1]])
```

```python
nodes=[]
edges=[]
df=pd.read_csv(src, usecols=['patent_id', 'citation_id'])
df.info()
```

```python
# for row in df:
#     nodes.append(row[0])
#     nodes.append(row[1])
#     edges.append((row[0],row[1]))
nodes=df['patent_id'].tolist()
nodes.extend(df['citation_id'].tolist())
```

```python
edges= list(zip(df.patent_id, df.citation_id))
```

```python
nodes=list(set(nodes)) #make unique
```

```python
print(len(nodes))
```

```python
print(len(edges))
```

```python
G = nx.DiGraph()
```

```python
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

```python
print(nx.info(G))
```

```python
katz=katz_centrality(G)
```
